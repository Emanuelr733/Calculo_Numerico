# Lista de Exercícios 02 – Raízes de Equações

## 1 - Plote o gráfico e encontre um intervalo contendo uma única raiz das seguintes funções:

```python
import numpy as np
import matplotlib.pyplot as pyplot

def plot(fx, a, b):
    lista_x = np.linspace(a, b)
    pyplot.plot(lista_x, fx(lista_x))
    pyplot.grid()
    pyplot.show()
```

### a) $f(x) = \ln x$

```python
def f(x):
    return np.log(x)

#plot(f,-10,10)
plot(f, 0.1, 2)
```

### b) $f(x) = e^x$

```python
def f(x):
    return np.exp(x)
plot(f,-100,-90)

# A função não toca no eixo x em momento nenhum
```

### c) $f(x) = \cos{(x)}$

```python
def f(x):
    return np.cos(x)
#plot(f,-10,10)
plot(f,0,2.5)
```

### d) $f(x) = x^2 + 2x + 1$

```python
def f(x):
    return (x**2 + 2*x + 1)
#plot(f,-10,10)
plot(f,-2,0)
```

### e) $f(x) = e^{(2 - x^2)} \times (x + 1)^2$

```python
def f(x):
    return (np.exp(2-x**2) * (x+1)**2)
#plot(f,-10,10)
#plot(f,-2,0)
plot(f,-1.5,-0.5)
```

### f) $f(x) = x^3 + 3x - 1$

```python
def f(x):
    return (x**3 + 3*x - 1)
#plot(f,-10,10)
plot(f,-2,2)
```

### g) $f(x) = x^2 + \sin{(x)}$

```python
def f(x):
    return (x**2 + np.sin(x))
#plot(f,-10,10)
#plot(f,-2,2)
plot(f,-0.5,0.5)
```

### h) $f(x) = e^{x} - x^2 - 10$

```python
def f(x):
    return (np.exp(x) -x**2 -10)
#plot(f,-10,10)
plot(f,-10,5)
```

### i) $f(x) = \sqrt{x} - \cos{(x)}$

```python
def f(x):
    return (np.sqrt(x) -np.cos(x))
#plot(f,10,10)
#plot(f,0,10)
plot(f,0,2)
```

### j) $f(x) = x - 3 - x^{-x}$

```python
def f(x):
    return (x-3-x**-x)
#plot(f,-10,10)
#plot(f,0,10)
plot(f,2,4)
```
