# Lista de Exercícios 03 – Raízes de Equações – Método da Bisseção

## 1 - Utilize o método da bisseção para encontre a raiz aproximada das seguintes equações. Considere os seguintes pontos:
- Precisão de 6 casas decimais;
- Número máximo de 50 iterações e $\epsilon < 10^{-5}$;
- Isole o intervalo da raiz;
- Analise a convergência do método;
- Apresente o código, solução e passo a passo para chegar na respostas.

```python
import numpy as np
import matplotlib.pyplot as pyplot
import pandas as pd

def plot(fx, a, b):
    lista_x = np.linspace(a, b)
    pyplot.plot(lista_x, fx(lista_x))
    pyplot.grid()
    pyplot.show()


def bissecao(f, a, b, tol, n):
    xant = float('nan')
    dados = pd.DataFrame()
    for i in range(n):
        x = (a + b) / 2
        fx = f(x)
        sinal = f(a) * fx
        error = abs((x-xant)/max(x, 1))
        xant = x
        dados_atual = pd.DataFrame({'a': [a], 'b': [b], 'erro': [error],
                                    'x': [x], 'f(x)': [fx], 'sinal': [sinal]})
        dados = pd.concat([dados, dados_atual], ignore_index=True)
        if (fx == 0) or (error < tol):
            break
        if sinal > 0:
            a = x
        else:
            b = x
    return x, dados
```

### a) $f(x) = \ln x$

```python
#definindo intervalo
def f(x):
    return np.log(x)

#plot(f, -10, 10)
plot(f, 0.1, 2)
raiz, dados = bissecao(f, 0.1, 2, 1e-5, 50)
print('Raiz encontrada: {r:+.6f}'.format(r=raiz))
display(dados)
```

### b) $f(x) = e^x$

```python
#definindo intervalo
def f(x):
    return np.exp(x)

plot(f, -100, 1)
#não tem raiz real
```

### c) $f(x) = \cos{(x)}$

```python
#definindo intervalo
def f(x):
    return np.cos(x)

#plot(f, -10, 10)
#plot(f, 1, 2)
plot(f, 1.5, 1.6)
raiz, dados = bissecao(f, 0, 2.5, 1e-5, 50)
print('Raiz encontrada: {r:+.6f}'.format(r=raiz))
display(dados)
```

### d) $f(x) = x^2 + 2x + 1$

```python
#definindo intervalo
def f(x):
    return x**2 + 2*x +1

plot(f, -1.1, -0.9)
raiz, dados = bissecao(f, -1.25, -0.75, 1e-5, 50)
print('Raiz encontrada: {r:+.6f}'.format(r=raiz))
display(dados)
```

### e) $f(x) = e^{(2 - x^2)} \times (x + 1)^2$

```python
#definindo intervalo
def f(x):
    return np.exp(2-x**2) * (x+1)**2

plot(f, -1.1, -0.9)
raiz, dados = bissecao(f, -1.1, -0.9, 1e-5, 50)
print('Raiz encontrada: {r:+.6f}'.format(r=raiz))
display(dados)
```

### f) $f(x) = x^3 + 3x - 1$

```python
#definindo intervalo
def f(x):
    return x**3 + 3*x - 1

#plot(f,-10,10)
plot(f,-1,1)
raiz, dados = bissecao(f, -1, 1, 1e-5, 50)
print('Raiz encontrada: {r:+.6f}'.format(r=raiz))
display(dados)
```

### g) $f(x) = x^2 + \sin{(x)}$

```python
#definindo intervalo
def f(x):
    return x**2 + np.sin(x)

#plot(f,-10,10)
#plot(f,-1,0)
plot(f,-1,-0.8)
raiz, dados = bissecao(f, -1, -0.8, 1e-5, 50)
print('Raiz encontrada: {r:+.6f}'.format(r=raiz))
display(dados)
```

### h) $f(x) = e^{x} - x^2 - 10$

```python
#definindo intervalo
def f(x):
    return np.exp(x) - x**2 -10

#plot(f,-10,10)
#plot(f,0,4)
plot(f,2.5,3)
raiz, dados = bissecao(f, 2.5, 3, 1e-5, 50)
print('Raiz encontrada: {r:+.6f}'.format(r=raiz))
display(dados)
```

### i) $f(x) = \sqrt{x} - \cos{(x)}$

```python
#definindo intervalo
def f(x):
    return np.sqrt(x) - np.cos(x)

#plot(f,-10,10)
plot(f,0,2)
raiz, dados = bissecao(f, 0, 2, 1e-5, 50)
print('Raiz encontrada: {r:+.6f}'.format(r=raiz))
display(dados)
```

```python
#definindo intervalo
def f(x):
    return x-3-x**-x

#plot(f,-10,10)
#plot(f,1,4)
plot(f,2,4)
raiz, dados = bissecao(f, 2, 4, 1e-5, 50)
print('Raiz encontrada: {r:+.6f}'.format(r=raiz))
display(dados)
```
