# Lista de Exercícios 04 – Raízes de Equações – Posição Falsa

## 1 - Utilize o método da posição falsa para encontre a raiz aproximada das seguintes equações. Considere os seguintes pontos:
- Precisão de 6 casas decimais;
- Número máximo de 50 iterações e $\epsilon < 10^{-5}$;
- Isole o intervalo da raiz;
- Analise a convergência do método;
- Apresente o código, solução e passo a passo para chegar na respostas.

```python
import numpy as np
import matplotlib.pyplot as pyplot
import pandas as pd

def plot(fx, a, b):
    lista_x = np.linspace(a, b)
    pyplot.plot(lista_x, fx(lista_x))
    pyplot.grid()
    pyplot.show()


def PosicaoFalsa(f, a, b, tol, n):
    xant = float('nan')
    dados = pd.DataFrame()
    for i in range(n):
        fa=f(a)
        fb=f(b)
        x = (a*fb - b*fa)/(fb-fa)
        fx=f(x)
        error = abs((x-xant)/max(x, 1))
        xant = x
        dados_atual = pd.DataFrame({'a': [a], 'b': [b], 'erro': [error],
                                    'x': [x], 'f(a)': [fa], 'f(b)': [fb], 'f(x)': [fx]})
        dados = pd.concat([dados, dados_atual], ignore_index=True)
        if (fx == 0) or (error < tol):
            break
        if fx * fa > 0:
            a = x
        else:
            b = x
    return x, dados
```

### a) $f(x) = \ln (x^2 + 1) - x$

```python
def f(x):
    return np.log(x**2 + 1) -x
#plot(f, -10, 10)
plot(f, -2.5, 2.5)
raiz, dados = PosicaoFalsa(f, -2.5, 2.5, 1e-6, 50)
print('Raiz encontrada: {r:+.6f}'.format(r=raiz))
display(dados)
```

### b) $f(x) = x \times \cos{(x)}$

```python
def f(x):
    return x * np.cos(x)
#plot(f, -10, 10)
plot(f, -1, 1)
raiz, dados = PosicaoFalsa(f, -1, 1, 1e-6, 50)
print('Raiz encontrada: {r:+.6f}'.format(r=raiz))
display(dados)
```

### c) $f(x) = 2x^6 -5x^3 + x - 1$

```python
def f(x):
    return 2*x**6 - 5*x**3 + x - 1
#plot(f, -10, 10)
#plot(f, -1, 1)
plot(f, -0.75, -0.5)
raiz, dados = PosicaoFalsa(f, -1, 1, 1e-6, 50)
print('Raiz encontrada: {r:+.6f}'.format(r=raiz))
display(dados)
```

### d) $f(x) = \dfrac{\tan(x)}{ln(x^2+1)}$

```python
def f(x):
    return np.tan(x)/np.log(x**2 + 1)
#plot(f, -10, 10)
#plot(f, 1, 2.5)
plot(f, 1.5, 1.6)
raiz, dados = PosicaoFalsa(f, 1.5, 1.6, 1e-6, 50)
print('Raiz encontrada: {r:+.6f}'.format(r=raiz))
display(dados)
```

### e) $f(x) = \frac{2x^3 - 4x + 1}{x^2 + 1}$

```python
def f(x):
    return (2*x**3 -4*x +1)/(x**2 + 1)
#plot(f, -10, 10)
plot(f, -2, -1)
#plot(f, 1.5, 1.6)
raiz, dados = PosicaoFalsa(f, -2, -1, 1e-6, 50)
print('Raiz encontrada: {r:+.6f}'.format(r=raiz))
display(dados)
```
